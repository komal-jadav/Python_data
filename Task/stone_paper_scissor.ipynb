```python
import random
#list for computer choice
Cchoice=["STONE","PAPER","SCISSOR"]
youwin,computerwin = 0,0
STATUS=True
while STATUS:
    print("STONE PAPER SCISSOR GAME:")
   
    

    
    print("ROUND START:")
    print("please select any one option:")
    print("1 for Stone","2 for Paper","3 for Scissor",sep="\n")
    Ychoice=int(input())
    if Ychoice==1:
        print("YOU SELECT STONE")
        Ychoice="STONE"
    elif Ychoice==2:
        print("YOU SELECT PAPER ")
        Ychoice="PAPER" 
    elif Ychoice==3:
        print("YOU SELECT SCISSOR")
        Ychoice="SCISSOR"

    else:    
        print("INVALID INPUT")
        
    Computer_choice=random.choice(Cchoice)
    print("COMPUTER SELECT:",Computer_choice)
    if Ychoice==Computer_choice:
        print("This Roun is Drawn:")
    elif(Ychoice=="PAPER" and Computer_choice=="STONE") or (Ychoice=="SCISSOR" and Computer_choice=="PAPER") or (Ychoice=="STONE" and Computer_choice=="SCISSOR"):
        
        print("YOU WIN THIS ROUND")
        youwin+=1
    else:
        
        print("COMPUTER WIN THIS ROUND")
        computerwin+=1
    user_choice= input("do you want to play again? press'y' for 'yes' press 'n' for 'no' ")
    if user_choice=='n'or user_choice=='no'or user_choice=='N' or user_choice=='No' or user_choice=="NO":
        STATUS=False
    else:
        STATUS=True    
if youwin>computerwin:
    print("YOU WIN THIS GAME")
    print("SCORE IS:","YOUR SCORE:",youwin,"COMPUTER SCORE:",computerwin)
elif computerwin>youwin:
    print("COMPUTER WIN THIS GAME")
    print("SCORE IS:","YOUR SCORE:",youwin,"COMPUTER SCORE:",computerwin)
else:
    print("MATCH DRAW")
    print("SCORE IS:","YOUR SCORE:",youwin,"COMPUTER SCORE:",computerwin)
    
    

    


```
